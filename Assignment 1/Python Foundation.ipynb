### 2. Hitunglah BMI setiap siswa dengan rumus
> BMI=Berat badan (kg) / (Tinggi Badan (m))^2





```python
berat_badan = df_g['Berat Badan (Kg)']
tinggi_badan = df_g['Tinggi Badan (Cm)']
bmi = berat_badan/((tinggi_badan * 0.01)**2)
print(bmi)
```

### 3. Buatlah program untuk menghitung BMI!

```python
def hitung_bmi(bb, tb):
  hasil = bb/((tb * 0.01)**2)
  return hasil

berat = float(input('Masukkan berat badan (Kg)  : '))
tinggi = float(input('Masukkan tinggi badan (Cm) : '))

hasil_bmi = hitung_bmi(berat, tinggi)
print('Nilai BMI: ', hasil_bmi)
```

### 4. Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
df_g['BMI'] = bmi
df_g
```

### 5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

  *   Normal: 18.5-24.9
  *   Overweight: 25.0-29.9
  *   Obesity: 30.0-34.9
  *   Underweight: < 18.5

```python
df_g.loc[(df_g['BMI'] >= 18.5) & (df_g['BMI'] <= 24.9), 'Status BMI'] = 'Normal'
df_g.loc[(df_g['BMI'] >= 25.0) & (df_g['BMI'] <= 29.9), 'Status BMI'] = 'Overweight'
df_g.loc[(df_g['BMI'] >= 30.0) & (df_g['BMI'] <= 34.9), 'Status BMI'] = 'Obesity'
df_g.loc[df_g['BMI'] <= 18.5, 'Status BMI'] = 'Underweight'

df_g
```
